```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sales_dataset.csv')
```

```python
df.head()
df.shape
df.columns
df.info()
df.describe()
df.isnull().sum()
```

**2. Data Cleaning**

```python
# Check missing values
df.isnull().sum()
```

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].median())
```

```python
df.drop(columns=['Unnamed: 22'], inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

**3. Data Analysis**

```python
df['Date'] = pd.to_datetime(df['Date'])
df.groupby(df['Date'].dt.to_period('M'))['Amount'].sum() #Revenue growth over time
```

Insights:Revenue shows a sharp increase from March to April 2022, followed by consistently high sales in May and June, indicating strong business scaling during this period.

```python
df.groupby('Category')['Amount'].sum().sort_values(ascending=False) #top selling
```

Insights:“Set” and “Kurta” dominate sales, contributing the majority of revenue, showing strong demand for ethnic wear products.

```python
df['Amount'].sum() / df['Order ID'].nunique() #average order value
```

Insights: The average order value of ₹692 indicates a moderately low basket size, suggesting high volume–low ticket sales strategy.

```python
df.groupby('ship-city')['Amount'].sum().sort_values(ascending=False).head(10) #Top locations
```

Insights: Bengaluru, Hyderabad, and Mumbai are the top revenue-generating cities, highlighting strong urban market concentration in metro cities.

```python
df[df['Status']=='Cancelled'].shape[0] / len(df) * 100 #%cancellation
```

Insights: A cancellation rate of 14.22% indicates moderate order leakage, suggesting scope for improving order confirmation and fulfillment processes.

```python
df.groupby('B2B')['Amount'].sum() / df['Amount'].sum() * 100 #b2b vs b2c
```

Insights:The business is highly B2C-driven (99.27%), showing strong dependence on individual customers rather than bulk business buyers.

```python
df['ship-service-level'].value_counts() #shipping performance
```

Insights: Expedited shipping is significantly higher than standard shipping, indicating customer preference for faster delivery services.

```python
df.groupby('Fulfilment')['Amount'].sum() / df['Amount'].sum() * 100 #amazon vs merchant
```

Insights: Amazon handles nearly 70% of orders, showing strong reliance on in-house fulfillment for operational efficiency.

```python
df.groupby('Category')['Amount'].sum().sort_values(ascending=False) #high demand categories
```

Insights: “Set” and “Kurta” remain the most demanded categories, confirming consistent customer preference for ethnic wear products.

```python
df['Qty'].mean() #quantity analysis
```

Insights: The average order quantity is low, indicating single-item purchase behavior dominates customer transactions.

```python
df['Status'].value_counts() #Return analysis
```

Insights: Most orders are successfully shipped and delivered, but cancellations and returns indicate operational inefficiencies and scope for improving customer satisfaction.
